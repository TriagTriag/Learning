```python
import warnings
warnings.filterwarnings("ignore") 
import torch as pt
import torch.nn as nn
import numpy as np
from matplotlib import pyplot as plt
from scipy.optimize import linprog 
N = 1000;
N_test = 200;
mu_1 = 1.0
mu_2 = 4.0
# pt.manual_seed(3091994)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
X1 = pt.randn([N,])+mu_1
Y1 = pt.ones([N,])
X2 = pt.randn([N,])+mu_2
Y2 = pt.zeros([N,])
X = pt.hstack([X1, X2])
Y = pt.hstack([Y1, Y2])
perm = pt.randperm(2*N)
X = X[perm].to(device)
Y = Y[perm].to(device)
Y_h = 1-Y
X1_test = pt.randn([N_test,])+mu_1
Y1_test = pt.ones([N_test,])
X2_test = pt.randn([N_test,])+mu_2
Y2_test = pt.zeros([N_test,])
X_test = pt.hstack([X1_test, X2_test])
Y_test = pt.hstack([Y1_test, Y2_test])
perm = pt.randperm(2*N_test)
X_test = X_test[perm].to(device)
Y_test = Y_test[perm].to(device)
Y_test_h = 1-Y_test
b = np.arange(0, 1, 0.05)
accs = np.zeros_like(b)
accs_order = np.zeros_like(b)
accs_thresh_adapt = np.zeros_like(b)
accs_thresh_adapt_offline = np.zeros_like(b)
accs_half = np.zeros_like(b)
accs_half_constrained = np.zeros_like(b)
exps = np.zeros_like(b)
exps_thresh_adapt = np.zeros_like(b)
exps_thresh_adapt_offline = np.zeros_like(b)
exps_half = np.zeros_like(b)
exps_half_constrained = np.zeros_like(b)
accs_mixed = np.zeros_like(b)
exps_mixed = np.zeros_like(b)
rnd_mixed  = np.zeros_like(b)
Tb = np.zeros_like(b)
Cb = np.zeros_like(b)
marg = np.zeros_like(b)
class LN(nn.Module):
    def __init__(self, dim_hidden):
        super().__init__()
        self.fc1 = nn.Linear(1, dim_hidden)
        pt.nn.init.xavier_uniform(self.fc1.weight)
        self.fc2 = nn.Linear(dim_hidden, 1)
        pt.nn.init.xavier_uniform(self.fc2.weight)
        self.relu = nn.ReLU();
        self.logsoftmax = nn.Sigmoid()
    def forward(self, in_):
        fc_out = self.relu(self.fc1(in_))
#         print(fc_out)
        out_1 = self.fc2(fc_out)
        out = self.logsoftmax(out_1)
#         print(out)
        return out
def train(model, X, Y, Y_h, b):
#     print(Y)
    X = X.detach()
    Y = Y.detach()
    Y_h = Y_h.detach()
#     print(model.parameters())
    chicken = 50
    num_epochs = 20
    num_batches = int(X.shape[0] / chicken)
    optimizer = pt.optim.Adam(model.parameters(), lr=0.03)
    loss_func = nn.BCELoss(reduction='none')
    train_losses = []
    best_val_loss = 1000
    eps = 1e-4
    
    for epoch in range(num_epochs):
#         print('----- epoch:',epoch, '-----')
        train_loss = 0
        machine_loss = []
        for i in range(num_batches):
            X_batch = X[i * chicken: (i + 1) * chicken]
            Y_batch = Y[i * chicken: (i + 1) * chicken]
            Y_h_batch = Y_h[i * chicken: (i + 1) * chicken]
            diff = Y_h_batch.detach().numpy()-Y_batch.detach().numpy()
            sol = linprog(diff,np.ones([1, chicken]), b*chicken, bounds = (0, 1))
            lab = sol.x
            lab = pt.tensor(lab, dtype = pt.float).unsqueeze(1)
            optimizer.zero_grad()
            loss = loss_func(model(X_batch),lab)
            loss.sum().backward()
            optimizer.step()
            train_loss += float(loss.mean())
        train_losses.append(train_loss / num_batches)
#         print('machine_loss:', train_loss/num_batches)

def test(model, X, Y, Y_H):
    h = model(X).detach().numpy()
    y = Y.detach().numpy()
    y_h = Y_H.detach().numpy()
#     print(h, y, y_h)
    acc = 1-np.sum(h*y_h+(1-h)*y)/len(Y)
    exp = np.sum(h)/len(Y)
#     print("Accuracy = ", acc)
#     print("Expected Deferral = ", exp)
    return acc, exp
def find_machine_samples(machine_loss, hloss,constraint):
    
    diff = machine_loss - hloss
    argsorted_diff = pt.clone(pt.argsort(diff))
#     print(argsorted_diff)
    num_outsource = int(constraint * machine_loss.shape[0])
#     print(constraint * machine_loss.shape[0])
    index = -num_outsource

    while (index < -1 and diff[argsorted_diff[index]] <= 0):
        index += 1

    
    if index==0:
        index = -1
    if index == -diff.shape[0]:
        index = 1
        
    machine_list = argsorted_diff[:index]

    return machine_list

def train_sorted(model, X, Y, Y_h, b):
#     print(Y)
    X = X.detach()
    Y = Y.detach()
    Y_h = Y_h.detach()
    length_X = X.shape[0]
    valid_perc = 0.2
    valid_num = int(np.floor(valid_perc*length_X))
    X_val = X[:valid_num]
    Y_val = Y[:valid_num].detach().numpy()
    Y_h_val = Y_h[:valid_num].detach().numpy()
    X = X[valid_num:]
    Y = Y[valid_num:]
    Y_h = Y_h[valid_num:]
    
    
#     print(model.parameters())
    chicken = 50
    num_epochs = 20
    num_batches = int(X.shape[0] / chicken)
    optimizer = pt.optim.Adam(model.parameters(), lr=0.03)
    loss_func = nn.BCELoss(reduction='none')
    train_losses = []
    best_val_loss = 1000
    eps = 1e-4
    
    for epoch in range(num_epochs):
#         print('----- epoch:',epoch, '-----')
        train_loss = 0
#         with pt.no_grad():
#             mprim = copy.deepcopy(model)
        machine_loss = []
        for i in range(num_batches):
            X_batch = X[i * chicken: (i + 1) * chicken]
            Y_batch = Y[i * chicken: (i + 1) * chicken]
            Y_h_batch = Y_h[i * chicken: (i + 1) * chicken]
#             diff = Y_h_batch.detach().numpy()-Y_batch.detach().numpy()
#             print(diff)
#             sol = linprog(diff,np.ones([1, chicken]), b*chicken, bounds = (0, 1))
#             lab = sol.x
# #             print(lab)
#             lab = pt.tensor(lab, dtype = pt.float).unsqueeze(1)
#             loss_temp = loss_func(model(X_batch), lab)
#             print(Y_batch, model(X_batch), loss_temp)
            machine_indices = find_machine_samples(Y_batch[:, 0], Y_h_batch[:, 0], b)
#             print(machine_indices)
            lab = pt.tensor([0 if j in machine_indices else 1 for j in range(Y_h_batch.shape[0])], dtype=pt.float)
            optimizer.zero_grad()
            loss = loss_func(model(X_batch),lab.unsqueeze(1))
#             print(Y_batch)
#             print(pt.autograd.grad(loss_temp.sum(), model.parameters()))
#             print(model(X_batch), Y_batch, loss)
            loss.sum().backward()
            optimizer.step()
            train_loss += float(loss.mean())
#             print(loss.mean())
        train_losses.append(train_loss / num_batches)
#         print('machine_loss:', train_loss/num_batches)

    
    num_machine = int(np.floor((1-b)*len(Y_val)))
    pred_val = model(X_val).T[0, :]
    human_candidate = pt.argsort(pred_val)[num_machine:]
    val_loss = np.zeros([valid_num-num_machine,])
    
    if (b!=0):
        for i in range(valid_num-num_machine):
            val_loss[i] = np.sum(Y_val[human_candidate[:i+1]]-Y_h_val[human_candidate[:i+1]])

        thresh = pred_val[human_candidate[np.argmin(val_loss)]]
        optim_i = np.argmin(val_loss)
    else:
        thresh = pt.tensor([1.0])
        optim_i = valid_num
    print("Threshold=", thresh)
    print("optimal i is ",optim_i)
    plt.plot(val_loss)
    return float(thresh.detach())
# print(Y)
# for name, param in model.named_parameters():
#     if param.requires_grad:
#         print (name, param.data)

# model(X.unsqueeze(1))
```

```python
def test_by_ordering(model, X, Y, Y_H, b, model_thresh):
    num_machine = int(np.floor((1-b)*len(Y)))
    thresh = np.arange(0.1, 1.1, 0.1)
    h = model(X).detach().numpy()
    y = Y.detach().numpy()
    y_h = Y_H.detach().numpy()
    h_to_sort = model(X).T[0, :]
#     print(h_to_sort.shape)
    to_machine = pt.argsort(h_to_sort)[:num_machine].cpu().data.numpy()
#     print("to_machine"+str(y[to_machine]))
    to_human = np.array([i for i in range(X.shape[0]) if i not in to_machine])
    if (len(to_human)==0):
        to_human = []
        
#     machine_candidates_adapt_thresh = pt.argsort(h_to_sort)[:num_machine].cpu().data.numpy()
#     print("to_machine"+str(y[to_machine]))
    to_human_adapt_thresh = np.array([i for i in range(X.shape[0]) if ( h_to_sort[i]>model_thresh)])
    to_machine_adapt_thresh = np.array([i for i in range(X.shape[0]) if i not in to_human_adapt_thresh])
    if (len(to_human_adapt_thresh)==0):
        to_human_adapt_thresh = []
    if (len(to_machine_adapt_thresh)==0):
        to_machine_adapt_thresh = []
        
        
        
    machine_candidates_adapt_thresh_offline = pt.argsort(h_to_sort)[:num_machine].cpu().data.numpy()
#     print("to_machine"+str(y[to_machine]))
    to_human_adapt_thresh_offline = np.array([i for i in range(X.shape[0]) if (i not in machine_candidates_adapt_thresh_offline and h_to_sort[i]>model_thresh)])
    to_machine_adapt_thresh_offline = np.array([i for i in range(X.shape[0]) if i not in to_human_adapt_thresh])
    if (len(to_human_adapt_thresh_offline)==0):
        to_human_adapt_thresh_offline = []
    if (len(to_machine_adapt_thresh_offline)==0):
        to_machine_adapt_thresh_offline = []
#     print(to_human, to_machine, to_human_adapt_thresh, to_machine_adapt_thresh)
#     print(to_human, to_machine)
    to_machine_half = pt.nonzero(h_to_sort<0.5)[:, 0].cpu().data.numpy()
    to_human_half = np.array([i for i in range(X.shape[0]) if i not in to_machine_half])
    if (len(to_human_half)==0):
        to_human_half = []
        
    to_machine_thresh = []
    to_human_thresh = []
    acc_thresh = []
    exp_thresh = []
    for i in range(len(thresh)):
        to_machine_thresh.append(pt.nonzero(h_to_sort<1-thresh[i])[:, 0].cpu().data.numpy())
        to_human_thresh.append(np.array([j for j in range(X.shape[0]) if j not in to_machine_thresh[i]]))
        if (len(to_human_thresh[i])==0):
            to_human_thresh[i] = []
        acc_thresh.append(1-np.sum(y_h[to_human_thresh[i]])/len(Y)-np.sum(y[to_machine_thresh[i]])/len(Y))
        exp_thresh.append(len(to_human_thresh[i])/len(Y))
        
        
    

    if len(Y)-num_machine<len(to_human_half):
        to_human_half_constrained = pt.argsort(h_to_sort[to_human_half])[num_machine-len(Y):].cpu().data.numpy()
    else:
        to_human_half_constrained = to_human_half
#     print(len(to_human_half_constrained))

#     print(h_to_sort[pt.argsort(h_to_sort)])
#     print(to_machine, to_machine_half)
    to_machine_half_constrained = np.array([i for i in range(X.shape[0]) if i not in to_human_half_constrained])
    if (len(to_machine_half_constrained)==0):
            to_machine_half_constrained = []
#     if (len(to_human_half_constrained)==0):
#         to_human_half_constrained = []
#     print("to_human_half"+str(y_h[to_human_half]))
#     print(h, y, y_h)
    acc = 1-np.sum(y_h[to_human])/len(Y)-np.sum(y[to_machine])/len(Y)
    acc_thresh_adapt = 1-np.sum(y_h[to_human_adapt_thresh])/len(Y)-np.sum(y[to_machine_adapt_thresh])/len(Y)
    acc_thresh_adapt_offline = 1-np.sum(y_h[to_human_adapt_thresh_offline])/len(Y)-np.sum(y[to_machine_adapt_thresh_offline])/len(Y)
    acc_half = 1-np.sum(y_h[to_human_half])/len(Y)-np.sum(y[to_machine_half])/len(Y)
    acc_half_constrained = 1-np.sum(y_h[to_human_half_constrained])/len(Y)-np.sum(y[to_machine_half_constrained])/len(Y)
#     print("Accuracy = ", acc, acc_half)
    exp_thresh_adapt = len(to_human_adapt_thresh)/len(Y)
    exp_thresh_adapt_offline = len(to_human_adapt_thresh_offline)/len(Y)
    exp_half = len(to_human_half)/len(Y)
    exp_half_constrained = len(to_human_half_constrained)/len(Y)
    return acc, acc_thresh_adapt, acc_thresh_adapt_offline, acc_half, acc_half_constrained, exp_thresh_adapt, exp_thresh_adapt_offline, exp_half, exp_half_constrained, acc_thresh, exp_thresh
```

```python
models = []
accs_thresh = []
exps_thresh = []
thresh_model = []
for i in range(len(b)):
    model = LN(10)
#     model_ = LN(10)
    thr = train_sorted(model, X.unsqueeze(1), Y.unsqueeze(1), Y_h.unsqueeze(1), b[i])
    thresh_model.append(thr)
    accs[i], exps[i] = test(model, X_test.unsqueeze(1), Y_test.unsqueeze(1), Y_test_h.unsqueeze(1))
    accs_order[i], accs_thresh_adapt[i], accs_thresh_adapt_offline[i], accs_half[i], accs_half_constrained[i], exps_thresh_adapt[i], exps_thresh_adapt_offline[i], exps_half[i], exps_half_constrained[i], acc_thresh, exp_thresh = test_by_ordering(model, X_test.unsqueeze(1), Y_test.unsqueeze(1), Y_test_h.unsqueeze(1), b[i], thresh_model[i])
    models.append(model)
    accs_thresh.append(acc_thresh)
    exps_thresh.append(exp_thresh)
```

```python
def validation_test(X, Y, Y_h, X_val, Y_val, Y_h_val, N_iter, bi, thr):
    accs_mixed_val = 0
    for j in range(N_iter):
        model1 = LN(10)
        model2 = LN(10)
        #     model_ = LN(10)
        Tb_val, Cb_val, marg_val = train_losses(model1, model2, X.unsqueeze(1), Y.unsqueeze(1), Y_h.unsqueeze(1), bi, thr, False)
        accs, exps_mixed_val, rnd_mixed_val = test_losses(model1, model2, X_val.unsqueeze(1), Y_val.unsqueeze(1), Y_h_val.unsqueeze(1), Tb_val, Cb_val, marg_val)
        accs_mixed_val += accs
    accs_mixed_val /=N_iter
    return accs_mixed_val

def point1(a, b, alpha):
    ret = (1+alpha)/2*a+(1-alpha)/2*b
    if ((ret - a)*( b-a)>np.abs(b-a)**2+1e-6):
        raise NameError('Err!')
    return (1+alpha)/2*a+(1-alpha)/2*b
def point2(a, b, alpha):
    ret = (1-alpha)/2*a+(1+alpha)/2*b
    if ((ret - a)*( b-a)>np.abs(b-a)**2+1e-6):
        raise NameError('Err!')
    return (1-alpha)/2*a+(1+alpha)/2*b

def golden (func, x0, v, MAX_ITER):
    class solution:
        def __init__(self, point, func, err, step, vector, threshold):
            self.maximizer = point
            self.eval_max = func
            self.err = err
            self.step = step
            self.opt_vec = vector
            self.threshold = threshold
#     if (pt.is_tensor(x0) ==0 or pt.is_tensor(v) == 0):
#         raise NameError('All the input points of function golden must be tensor.')

    #parameters
    alpha = np.sqrt(5, dtype = np.double)-2
    ABSTOL = 1e-8
    
    #Defs
    a = np.zeros([MAX_ITER,])
    b = np.zeros([MAX_ITER,])
    x1 = np.zeros([MAX_ITER,])
    x2 = np.zeros([MAX_ITER,])
    prec_threshold = False
    
    #Initialization of a and b
    a[0] = x0
    b[0] = x0+v
    #Main loop
    for i in range(MAX_ITER-1):
        print('Iteration '+str(i))
        if (np.abs(a[i]- b[i])<=np.abs(a[0]- b[0])/100):
            prec_threshold = True
            break;
        if (i==0):
            x1[i] = point1(a[i], b[i], alpha)
            x2[i] = point2(a[i], b[i], alpha)
        f1 = func(x1[i])
        f2 = func(x2[i])
        print(a[i], x1[i], x2[i], b[i])
#         print('f1, f2, f3, f4 =', func(a[i, :]), func(x1[i, :]), func(x2[i, :]), func(b[i, :]))
#         print('Err =', pt.abs(func(x1[i, :])-func(x2[i, :])).detach().numpy()<ep_diff)
        if (np.abs(f1-f2)<ABSTOL):
            print('Case I')
            a[i+1] = x1[i]
            b[i+1] = x2[i]
            x1[i+1] = point1(a[i+1], b[i+1], alpha)
            x2[i+1] = point2(a[i+1], b[i+1], alpha)
        elif (f1<f2):
            print('Case II')
            a[i+1] = x1[i]
            b[i+1] = b[i]
            x1[i+1] = x2[i]
            x2[i+1] = point2(a[i+1], b[i+1], alpha)
        elif (f1>f2):
            print('Case III')
            a[i+1] = a[i]
            b[i+1] = x2[i]
            x2[i+1] = x1[i]
            x1[i+1] = point1(a[i+1], b[i+1], alpha)
        if (np.abs(a[i]-b[i])-np.abs(a[0]-b[0])>0):
                raise NameError('Golden Search is going outside of the iterval! ||a-b||_2 = '+str(np.abs(a[0]-b[0]))+' ||a_k - b_k||_2 = '+str(np.abs(a[i]-b[i]).detach().numpy()))
    maximizer = (a[i]+b[i])/2 # i takes MAX_ITER - 1 in normal loop without break
#     print(pt.matmul(minimizer-x0, v.T))
#     print('v = ',v)
    err = b[i]-a[i]
    step = np.abs(maximizer-x0)/np.abs(v)
    print('step =' +str(step))
#             print('Grad_f_new.p = '+str(pt.matmul(grad_func(x0+v*step), v.T).detach().numpy()))
    if ((func(maximizer)-func(x0))<-ABSTOL):
        print('Returned '+str(maximizer)+' but the result wasn\'t increasing. Indeed, the evaluation of function at the obtained point was '+ str(func(maximizer))+' whereas the x0 function was '+str(func(x0))+'.')
        print('||v||_2 = ', str(np.abs(v)))
        print('||maximizer - x0||_2 = ', str(np.abs(maximizer-x0)))
        return golden (func, x0, v/2, MAX_ITER)
    else:
        print(func(maximizer))
        return solution(maximizer, func(maximizer), err, step, v, prec_threshold)

def train_losses(model1, model2, X, Y, Y_h, b, margin, adaptive_margin=True):
    TOL =  1e-5
#     margin = 0.01
#     print(Y)
    X = X.detach()
    Y = Y.detach()
    Y_h = Y_h.detach()
    if (adaptive_margin==True):
        length_X = X.shape[0]
        valid_perc = 0.2
        valid_num = int(np.floor(valid_perc*length_X))
        X_val = X[:valid_num]
        Y_val = Y[:valid_num]
        Y_h_val = Y_h[:valid_num]
        X = X[valid_num:]
        Y = Y[valid_num:]
        Y_h = Y_h[valid_num:]
        valid_fun = lambda tau: validation_test(X, Y, Y_h, X_val, Y_val, Y_h_val, 10, b, tau)
        sol = golden(valid_fun, 0, 0.1, 10)
        margin = sol.maximizer
#     print(model.parameters())
    chicken = 50
    num_epochs = 20
    num_batches = int(X.shape[0] / chicken)
    optimizer1 = pt.optim.Adam(model1.parameters(), lr=0.03)
    optimizer2 = pt.optim.Adam(model2.parameters(), lr=0.03)
    loss_func = nn.BCELoss(reduction='none')
    train_losses = []
    best_val_loss = 1000
    eps = 1e-4
    
    for epoch in range(num_epochs):
#         print('----- epoch:',epoch, '-----')
        train_loss = 0
#         with pt.no_grad():
#             mprim = copy.deepcopy(model)
        machine_loss = []
        for i in range(num_batches):
            X_batch = X[i * chicken: (i + 1) * chicken]
            Y_batch = Y[i * chicken: (i + 1) * chicken]
            optimizer1.zero_grad()
            loss1 = loss_func(model1(X_batch),Y_batch)
#             print(Y_batch)
#             print(pt.autograd.grad(loss_temp.sum(), model.parameters()))
#             print(model(X_batch), Y_batch, loss)
            loss1.sum().backward()
            optimizer1.step()
            train_loss += float(loss1.mean())
#             print(loss.mean())
        train_losses.append(train_loss / num_batches)
#         print('machine_loss:', train_loss/num_batches)

    for epoch in range(num_epochs):
#         print('----- epoch:',epoch, '-----')
        train_loss = 0
#         with pt.no_grad():
#             mprim = copy.deepcopy(model)
        machine_loss = []
        for i in range(num_batches):
            X_batch = X[i * chicken: (i + 1) * chicken]
            Y_h_batch = Y_h[i * chicken: (i + 1) * chicken]
            optimizer2.zero_grad()
            loss2 = loss_func(model2(X_batch),Y_h_batch)
#             print(Y_batch)
#             print(pt.autograd.grad(loss_temp.sum(), model.parameters()))
#             print(model(X_batch), Y_batch, loss)
            loss2.sum().backward()
            optimizer2.step()
            train_loss += float(loss2.mean())
#             print(loss.mean())
        train_losses.append(train_loss / num_batches)
#         print('machine_loss:', train_loss/num_batches)
    l_AI = model1(X)
    l_H  = model2(X)
    d = (l_H-l_AI).detach().numpy()
    neg_num = np.sum(d[:, 0]<=TOL)
    if (b>=neg_num/X.shape[0]):
#         print('After cut-off!')
        Tb = 0
        Cb = 0
    else:
#         print('Before cut-off!')
#         print(int(np.floor(b*X.shape[0])))
        Tb = np.sort(d[:,0])[int(np.floor(b*X.shape[0]))]
        u = np.argsort(X[:, 0])
#         plt.plot(X[u].detach().numpy(), d[u])
        ind_in_marg = np.logical_and(d[:,0]>=(Tb-margin), d[:, 0]<=(Tb+margin))
#         print(ind_in_marg.shape)
        ind_more = d[:, 0]>(Tb+margin)
        ind_less = d[:, 0]<(Tb-margin)
        num_ds_in_marg = np.sum(ind_in_marg)
        num_ind_less = np.sum(ind_less)
#         print(num_ds_in_marg, num_ind_less)
        Cb = (b*X.shape[0]-num_ind_less)/num_ds_in_marg
#     print('Finished training!')
#     print(b, Tb, Cb, margin)
    return Tb, Cb, margin

def test_losses(model1, model2, X, Y, Y_H, Tb, Cb, margin):
    TOL=1e-5
    l_AI = model1(X).detach().numpy()
    l_H = model2(X).detach().numpy()
    d = (l_H-l_AI)
    h1 = (d[:, 0]<(Tb-margin))*1.0
    h2 = (np.logical_and(d[:,0]>=(Tb-margin), d[:, 0]<=(Tb+margin)))*Cb
    h = h1+h2
#     print(h)
    y = Y.detach().numpy()
    y_h = Y_H.detach().numpy()
#     print(h, y, y_h)
    acc = 1-np.sum(h*y_h[:,0]+(1-h)*y[:,0])/len(Y)
    exp = np.sum(h)/len(Y)
    rnd = np.sum((np.logical_and(d[:,0]>=(Tb-margin), d[:, 0]<=(Tb+margin)))*1.0)/(len(Y))
#     print("Accuracy = ", acc)
#     print("Expected Deferral = ", exp)
    return acc, exp, rnd
```
